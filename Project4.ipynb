```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('network_backup_dataset.csv')
dataset.head()
```

```python
dataset
```

```python
workflows = dataset['Work-Flow-ID'].unique()
workflows
```

```python
dataset['File Name'].unique()
```

```python
import calendar
def encode_day_names(days):
    day_to_num = dict(zip(list(calendar.day_name), range(1, 8)))
    return [day_to_num[day] for day in days]
```

```python
dataset['Day #'] = (dataset['Week #'] - 1) * 7 + encode_day_names(dataset['Day of Week'])
dataset
```

```python
first20 = dataset.loc[dataset['Day #'] <= 20]
first20
```

```python
def plot_dataset(df):
    for wid, wgrp in df.groupby('Work-Flow-ID'):
        fig, ax = plt.subplots()
        labels = []

        for fid, fgrp in wgrp.groupby('File Name'):
            ax = fgrp.groupby('Day #').sum().plot(
                ax = ax,
                kind = 'line',
                y = 'Size of Backup (GB)',
            )
            labels.append(fid)

        lines, _ = ax.get_legend_handles_labels()

        plt.xlabel('Time Period (days)')
        plt.ylabel('Data copy size (GB)')

        ax.legend(lines, labels, loc='best')
        ax.set_title('Copy size variance for ' + wid)
```

```python
plot_dataset(first20)
```

```python
first105 = dataset.loc[dataset['Day #'] <= 105]
```

```python
plot_dataset(first105)
```

```python
def plot_dataset2(df):
    fig, ax = plt.subplots()
    labels = []
    for wid, wgrp in df.groupby('Work-Flow-ID'):
        ax = wgrp.groupby('Day #').sum().plot(
                ax = ax,
                kind = 'line',
                y = 'Size of Backup (GB)',
            )
        labels.append(wid)

        lines, _ = ax.get_legend_handles_labels()

        plt.xlabel('Time Period (days)')
        plt.ylabel('Data copy size (GB)')

        ax.legend(lines, labels, loc='best')
        ax.set_title('Copy size variance')
```

```python
plot_dataset2(first20)
```
